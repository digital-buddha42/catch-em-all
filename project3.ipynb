We have 1017 different pokemon, we need to loop through and collect all of the names

```python
names_list =[]
for i, result in enumerate(results):
#     print (results[i]['name'])
    names_list.append(results[i]['name'])
    
names_list
```

```python
# Set the API base URL
url = 'https://pokeapi.co/api/v2/pokemon/'

# Define an empty list to fetch the data for each pokemon
pokemon_data = []

# Loop through all the pokemon in our list to fetch data
for i, name in enumerate(names_list):
        

    # Create endpoint URL with each pokemon
    pokemon_url = url + name
    
    pokemon_response = requests.get(pokemon_url)
    pokemon_json = pokemon_response.json()

    abilities = pokemon_json['abilities'][0]['ability']['name']
    base_experience = pokemon_json['base_experience']
    height = pokemon_json['height']
    id = pokemon_json['id']
    types = pokemon_json['types'][0]['type']['name']
    
    try: 
        type_2 = pokemon_json['types'][1]['type']['name']
        
    except:
        type_2 = 'No second type'
    
    weight = pokemon_json['weight']
    hp = pokemon_json['stats'][0]['base_stat']
    attack = pokemon_json['stats'][1]['base_stat']
    defense = pokemon_json['stats'][2]['base_stat']
    special_attack = pokemon_json['stats'][3]['base_stat']
    special_defense = pokemon_json['stats'][4]['base_stat']
    speed = pokemon_json['stats'][5]['base_stat']
    sprites = pokemon_json['sprites']['other']['official-artwork']['front_default']

    # Append the Pokemon stats information into city_data list
    pokemon_data.append({"Pokemon" : name,
                         "Abilities": abilities, 
                         "Base_Experience": base_experience, 
                         "Height": height, 
                         "ID": id,
                         "Type_1": types,
                         "Type_2" : type_2,
                         "Weight": weight,
                         "HP": hp,
                         "Attack": attack,
                         "Defense": defense,
                         "Special_Attack" : special_attack,
                         "Special_Defense": special_defense,
                         "Speed": speed,
                         "Sprites": sprites})
```

```python
# Convert the pokemon data into a Pandas DataFrame
poke_data_df = pd.DataFrame(pokemon_data)

# Show Record Count
print(poke_data_df)
print(poke_data_df.count())
```

```python
filter_poke_df = poke_data_df.dropna()
print(filter_poke_df)
print(filter_poke_df.count())
```

## Add Evolution Data in seperate API call

```python
# Set the API base URL for Evolution endpoint
url_2 = 'https://pokeapi.co/api/v2/evolution-chain/'

# Define an empty list to fetch the data for each evolution chain
evolution_data = []

for i in range(544):
    
    # Create endpoint URL with each pokemon
    evolution_url = url_2 + str(i + 1)
    
    evolution_response = requests.get(evolution_url)
    
    if not evolution_response.text.strip():  # Check if response content is empty
        print(f"Response content for {evolution_url} is empty. Skipping to the next iteration.")
        continue  # Skip to the next iteration
    
    try:
        evolution_json = evolution_response.json()  
    except json.JSONDecodeError as e:
        
        print(f"Error decoding JSON: No results for id: {i+1}")
        continue  # Skip to the next iteration
    
    # Obtain evolution chain data and store in a list
    base_pokemon = evolution_json['chain']['species']['name']
    
    try:
        evolves_to_1 = evolution_json['chain']['evolves_to'][0]['species']['name']
    except:
        evolves_to_1 = 'No first evolution'
    
    try:
        evolves_to_2 = evolution_json['chain']['evolves_to'][0]['evolves_to'][0]['species']['name']
    except:
        evolves_to_2 = 'No second evolution'

    evolution_chain = [base_pokemon, evolves_to_1, evolves_to_2]
    
    # Append the evolution chain as key-value pairs with all levels of evolution
    evolution_data.append({"Base_Pokemon": base_pokemon,
                           "Evolution_Level_1": evolves_to_1,
                           "Evolution_Level_2": evolves_to_2})
```

```python
evolution_data_df =  pd.DataFrame(evolution_data)
evolution_data_df
```

```python
base_join_df = pd.merge(filter_poke_df, evolution_data_df, left_on='Pokemon', right_on='Base_Pokemon', 
                     how='left')

ev_1_join_df = pd.merge(base_join_df, evolution_data_df, left_on='Pokemon', right_on='Evolution_Level_1', 
                     how='left')

ev_2_join_df = pd.merge(ev_1_join_df, evolution_data_df, left_on='Pokemon', right_on='Evolution_Level_2', 
                     how='left')

ev_2_join_df['Base_Pokemon'].fillna(ev_2_join_df['Base_Pokemon_x'], inplace=True)
ev_2_join_df['Base_Pokemon'].fillna(ev_2_join_df['Base_Pokemon_y'], inplace=True)

ev_2_join_df['Evolution_Level_1'].fillna(ev_2_join_df['Evolution_Level_1_x'], inplace=True)
ev_2_join_df['Evolution_Level_1'].fillna(ev_2_join_df['Evolution_Level_1_y'], inplace=True)

ev_2_join_df['Evolution_Level_2'].fillna(ev_2_join_df['Evolution_Level_2_x'], inplace=True)
ev_2_join_df['Evolution_Level_2'].fillna(ev_2_join_df['Evolution_Level_2_y'], inplace=True)

columns_to_delete = ['Base_Pokemon_x','Evolution_Level_1_x','Evolution_Level_2_x','Base_Pokemon_y',
                     'Evolution_Level_1_y','Evolution_Level_2_y']

ev_2_join_df.drop(columns=columns_to_delete, inplace=True)

ev_2_join_df.head()
```

```python
ev_2_join_df['Pokemon'] = ev_2_join_df['Pokemon'].str.title()
ev_2_join_df['Abilities'] = ev_2_join_df['Abilities'].str.title()
ev_2_join_df['Type_1'] = ev_2_join_df['Type_1'].str.title()
ev_2_join_df['Type_2'] = ev_2_join_df['Type_2'].str.title()
ev_2_join_df['Base_Pokemon'] = ev_2_join_df['Base_Pokemon'].str.title()
ev_2_join_df['Evolution_Level_1'] = ev_2_join_df['Evolution_Level_1'].str.title()
ev_2_join_df['Evolution_Level_2'] = ev_2_join_df['Evolution_Level_2'].str.title()

ev_2_join_df['Base_Pokemon'] = ev_2_join_df['Base_Pokemon'].fillna('No Evolution Chain')
ev_2_join_df['Evolution_Level_1'] = ev_2_join_df['Evolution_Level_1'].fillna('No Evolution Chain')
ev_2_join_df['Evolution_Level_2'] = ev_2_join_df['Evolution_Level_2'].fillna('No Evolution Chain')

ev_2_join_df.head()
```

```python
filter_poke_merged_df = ev_2_join_df

#Create directory for CSV files and loop through data to write CSV files
if not os.path.exists("output"):
    os.makedirs("output")
    
csv_filename = 'pokemon_data.csv'
file_path = os.path.join("output", csv_filename)


filter_poke_merged_df.to_csv(file_path, index=False)
```

```python
#Write each CSV file into SQLite table
database_path = "pokemon.sqlite"
Path(database_path).touch()

conn = sqlite3.connect(database_path)
c = conn.cursor()

for filename in os.listdir("output"):
    if filename.endswith(".csv"):
        table_name = os.path.splitext(filename)[0]

        c.execute(f'''CREATE TABLE IF NOT EXISTS {table_name} (
                        Pokemon TEXT,
                         Abilities TEXT, 
                         'Base_Experience' INT, 
                         Height INT, 
                         ID INT,
                         'Type_1' TEXT,
                         'Type_2' TEXT,
                         Weight INT,
                         HP INT,
                         Attack INT,
                         Defense INT,
                         'Special_Attack' INT,
                         'Special_Defense' INT,
                         Speed INT,
                         'Base_Pokemon' TEXT,
                         'Evolution_Level_1' TEXT,
                         'Evolution_Level_2' TEXT,
                         Sprites TEXT
                     )''')

        csv_data = pd.read_csv(os.path.join("output", filename))
        csv_data.to_sql(table_name, conn, if_exists='replace', index=False)

conn.commit()
conn.close()
```
